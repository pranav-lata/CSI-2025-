### Pie Plot

```python
df = pd.read_csv(r"Used_Bikes.csv")
df.head()
```

```python
df['owner'].value_counts()
```

```python
labels = list(df['owner'].value_counts().keys())
values = list(df['owner'].value_counts().values)
```

```python
plt.pie(values,labels=labels)
plt.show()
```

```python
#Population
population = ["India","China","Japan","USA"]
population_count = [27000,24000,11000,20000]
ls = [0,0,0.2,0]
plt.pie(population_count,labels=population,autopct='%1.2f%%',wedgeprops = {'ec':'k'},explode=ls,shadow=True)
plt.show()
```

```python
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
tip.columns
```

```python
#Renaming Columns
tip.columns = ['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

```python
tip.head()
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tip)
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip)
plt.show()
# By default it will create scatter plot
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,kind='line')
plt.show()
```

```python
tip
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker',col='gender')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker',row='gender')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='gender',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='time',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=tip,hue='time',style='smoker',col='day',kind='line')
plt.show()
```
